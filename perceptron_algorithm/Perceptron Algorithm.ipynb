# Perceptron Algorithm

Using the perceptron algorithm to find the proper weights for a linear classification


### Creating data

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# we will set the real line to be 3w_1 + 4w_2 - 10
# the slope-intercept form being: w_2 = (10 - 3w_1)/4

over = [(np.random.randint(0, 10), np.random.randint(0, 10)) for a in range(10)]
under = [(np.random.randint(-10, 0), np.random.randint(-10, 0)) for a in range(10)]

plt.scatter([a[0] for a in over], [a[1] for a in over], color="blue")
plt.scatter([a[0] for a in under], [a[1] for a in under], color="red")
plt.show()
```

### Finding the best classifying line for the data

```python
alpha = 1

W = np.array(np.random.randint(1, 4, 2)) # weights
b = np.random.randint(-10, 10) # bias

X = [np.array(asdf) for asdf in over+under]
y = [1 for o in over] + [0 for u in under]

iterations = 1000
adds = 0
subs = 0
for j in range(iterations):
    for i in range(len(X)):
        t = np.dot(W, X[i]) + b
        y_hat = 1 if t >= 0 else 0
        if y[i] - y_hat == -1:
            W = W + alpha*X[i]
            b = b + alpha
            adds += 1
        elif y[i] - y_hat == 1:
            W = W - alpha*X[i]
            b = b - alpha
            subs += 1
            
plt.scatter([a[0] for a in over], [a[1] for a in over], color="blue")
plt.scatter([a[0] for a in under], [a[1] for a in under], color="red")
plt.plot(range(-10, 10), [((-W[0]*x) - b)/W[1] for x in range(-10, 10)])
plt.show()
```
