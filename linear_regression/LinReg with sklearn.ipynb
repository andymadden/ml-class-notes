# Linear Regression Adventures

Working with the linear regression course

```python
%matplotlib inline
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

x = [a for a in range(100)]
y = [a + np.random.randint(-100, 100) for a in range(100)]

plt.scatter(x, y)
plt.show()
```

```python
lin_model = LinearRegression()
regression = lin_model.fit(np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1))

y_p = [regression.predict(np.array([[a]]))[0][0] for a in range(100)]

plt.scatter(x, y)
plt.plot(x, y_p)
plt.show()
```
